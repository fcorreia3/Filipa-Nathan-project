```python
import requests
from requests.exceptions import ConnectionError, Timeout, TooManyRedirects
import json
from config import api_key
import pandas as pd
from datetime import datetime
import pymongo
```

```python
exchanges_df = pd.read_csv("Resources/Exchanges.csv")
traffic_df = pd.read_csv("Resources/traffic_data.csv")
```

```python
df = pd.merge(exchanges_df,traffic_df, how = "right", left_on = "Url", right_on = "url")
```

```python
df = df.drop(columns = "Unnamed: 0_x")
df = df.drop(columns = "Unnamed: 0_y")
df = df.drop(columns = "url")

df = df.rename(columns={"Url": "URL", "rank": "alexa_rank"})
```
