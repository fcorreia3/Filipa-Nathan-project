```python
exchanges_usdt = df_btc_usdt['exchange']
echanges_usd = df_btc_usd['exchange']

exchanges = list(set(exchanges_usdt) | set(echanges_usd)) 

# exchanges = str(exchanges).replace("'", '"')
```

```python
exchanges
```

```python
exchange_list = []
times = []
volumes = []

for exchange in exchanges[:]:
    url = "https://min-api.cryptocompare.com/data/exchange/histoday?"

    payload = {
        "api_key": api_key,
        "tsym": "USD",
        "e": exchange,
        "limit" : 1
    }

    result = dict(requests.get(url, params=payload).json())
    result['exchange'] = exchange
    
    time = result['Data'][0]['time']
    times.append(time)

    volume = result['Data'][0]['volume']
    volumes.append(volume)
    
    exchange = result['exchange']
    exchange_list.append(exchange)
```

```python
d = {'time':times,'volume':volumes, 'exchange': exchange_list}
df_volumes = pd.DataFrame(d)
df_volumes

date = []
for unix in df_volumes['time']:
    date.append(datetime.utcfromtimestamp(unix).strftime('%Y-%m-%d'))
    
df_volumes['date'] = date
df_volumes.drop( columns = ['time'])
```

```python
url = "https://min-api.cryptocompare.com/data/exchanges/general"

payload = {
    "api_key": api_key,
}

result = dict(requests.get(url, params=payload).json())
result[0]
```

```python
list(result)
```

```python
list_ids = list(result['Data'])
```

```python
# for loop
# Id:
# Name:
# AffiliateURL:
df = pd.DataFrame(columns = ['Id', 'Name', 'Url'])

for i in range(len(list_ids)):
    output = result['Data'][list_ids[i]]
    output_id = output['Id']
    output_name = output['Name']
    output_url = output['AffiliateURL']
    current_df = pd.DataFrame({
        "Id":[str(output_id)],
        "exchange":[str(output_name)],
        "Url":[str(output_url)]
    })
    df = df.append(current_df)
#     print(output_id, output_name, output_url)
df
```

```python
df
```

```python
df_merged = pd.merge(df,df_volumes, how = 'inner', left_on= "Name", right_on = "exchange")
df_merged
```
